```python
## Funcion de Formateo de datos 
def formatData(data):
    data=str(data)
#     data=data.replace('< ','0')
#     data=data.replace('> ','0')
#     data=data.replace('<','0')
#     data=data.replace('>','0')
    if '<' in data or '>' in data or 'S' in data:
        data=np.NaN
        return data
    else:
        data=data.replace(',','.')
        data=data.replace('O ','0.')
        return float(data)
```
